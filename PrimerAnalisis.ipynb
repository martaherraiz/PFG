**Variables:**

PatientID: identificador único para cada paciente.

Age: Edad del paciente, todos son entre 60 y 90 años ya que es el grupo que interesa estudiar relacionado con esta enfermedad.

Gender: género del paciente donde 0 significa ser hombre y 1 significa ser mujer.

Ethnicity: etnia de cada paciente donde 0 es ser caucásico, 1 afroamericano, 2 asiático y 3 'otros'.

EducationLevel: el nivel de educación de cada paciente donde 0 es sin estudios, 1 es secundaria, 2 es licenciado y 3 son maestrías o doctorados.

BMI: índice de masa corporal entre 15 y 40, teniendo en cuenta que los valores entre los que debe estar una persona para estar sana son 18 y 29 aproximadamente.

Smoking: indica si el paciente es o no fumador, 0 indica que no y 1 que sí.

AlcoholConsumption: el alcohol que consume en una semana el paciente con unidades entre 0 y 20.

PhysicalActivity: son las horas semanales de actividad física entre 0 y 10, estos valores son más bajos que los de una persona joven, pero hay que tener en cuenta que son gente de más de 60 años.

DietQuality: la calidad de la dieta de cada paciente con puntuaciones entre 0 y 10.

SleepQuality: calidad de sueño de cada paciente con un rango entre 4 y 10.

FamilyHistoryAlzheimers: indica si el paciente tiene o no familiares que hayan tenido la enfermedad donde 0 indica que no y 1 que sí.

CardiovascularDisease: indica si el paciente tiene enfermedades cardiovasculares, 0 indica que no y 1 que sí.

Diabetes: indica si el paciente padece diabetes, 0 es que no y 1 que sí.

Depression: indica si el paciente padece depresión, 0 es que no y 1 que sí.

HeadInjury: indica si el paciente ha sufrido algún golpe en la cabeza, 0 es que no y 1 que sí.

Hypertension: indica si el paciente padece hipertension, en caso de que no es 0 y en caso de que sí es 1.

SystolicBP: presión arterial sistólica, que oscila entre 90 y 180 mmHg.

DiastolicBP: presión arterial diastólica, que oscila entre 60 y 120 mmHg.

*Diferencia: Presión sistólica: Mide la presión cuando el corazón bombea sangre (fase de contracción). Presión diastólica: Mide la presión cuando el corazón está en reposo (fase de relajación))


CholesterolTotal: Es el nivel total de colesterol en sangre, que incluye la suma del colesterol LDL, HDL y otras. Se mide en mg/dL y en la base de datos está entre 150 y 300 mg/dL donde los valores altos no son buenos.

CholesterolLDL: (colesterol malo) Es el nivel de lipoproteínas de baja densidad en la sangre. Estas lipoproteínas transportan colesterol hacia las arterias, lo que puede causar acumulación de placa en las paredes arteriales, aumentando el riesgo de enfermedades cardíacas. En la base se encuentra entre los valores 50 y 200 mg/dL donde los valores más elevados son los más peligrosos.

CholesterolHDL: (colesterol bueno) s el nivel de lipoproteínas de alta densidad en la sangre. Las HDL transportan el colesterol de las arterias al hígado para su eliminación, lo que ayuda a proteger contra enfermedades cardíacas. En la base se encuentra entre 20 y 100 mg/dL donde los valores más altos son beneficiosos.

CholesterolTriglycerides: Son un tipo de grasa en la sangre que el cuerpo utiliza como fuente de energía. Los triglicéridos elevados pueden estar relacionados con el aumento de enfermedades cardíacas. Se encuentra entre los valores 50 y 400 mg/dL donde los valores más altos son los más peligrosos.

MMSE (Mini-Mental State Examination): se utiliza para evaluar el estado cognitivo de un paciente, midiendo funciones como la orientación, atención, memoria y lenguaje. Los valores están entre 0 y 30 donde la puntuación más baja indica más deterioro.

FunctionalAssessment: mide la capacidad del paciente para realizar actividades cotidianas y funciones básicas. Los valores son entre 0 y 10 donde los valores más bajos indican un mayor deterioro.

MemoryComplaints: indica si el paciente reporta problemas o quejas relacionadas con la memoria. 0 indica que el paciente no tiene quejas y 1 que sí.

BehavioralProblems: indica si el paciente presenta problemas de comportamiento, que pueden incluir irritabilidad, agresividad, o cambios en el estado de ánimo. 0 indica que no y 1 que sí.

ADL: indica la capacidad del paciente para realizar actividades diarias básicas como vestirse, alimentarse y bañarse. Los valores están entre 0 y 10, donde los altos son los buenos.

Confusion: si el paciente presenta confusión, que puede manifestarse como falta de claridad en el pensamiento o en la percepción de la realidad. El 0 indica que no y el 1 que sí.

Disorientation: indica si el paciente está desorientado en tiempo, lugar o persona. 0 indica que no y 1 que sí.

PersonalityChanges: si hay cambios notables en la personalidad del paciente, como cambios en el comportamiento, el estado de ánimo o la forma de interactuar con otros. 0 indica que no y 1 que sí.

DifficultyCompletingTasks: si el paciente tiene problemas para llevar a cabo tareas cotidianas o actividades que solía realizar sin dificultad. 0 indica que no y 1 que sí.

Forgetfulness: i el paciente presenta episodios de olvido o pérdida de memoria, que pueden afectar su capacidad para recordar información reciente o pasados eventos. 0 indica que no y 1 que sí.

Diagnosis: es el diagnóstico que indica si el paciente tiene Alzheimer o no, 1 es que sí y 0 que no.

DoctorInCharge: es información confidencial sobre el médico a cargo del paciente. Esta columna no aporta nuevos datos al proyecto, por lo que será eliminada.

#Visualización de los datos:

Se hará una descripción gráfica de los datos para entender las distribuciones y detectar outliers, en caso de que los haya se decidirá más adelante que hacer con ellos.

Primero, se verá qué variables toman valores continuos, es decir se busca excluir aquellas que toman dos o tres valores únicos en la representación gráfica:

```python
unique_values = df.nunique()

print(unique_values)
```

Ya que se tienen las variables con más valores, se verán gráficamente mediante boxplots para examinar sus outliers:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Lista de variables a excluir en la representación
binary_vars = ['Gender', 'Ethnicity', 'EducationLevel', 'Smoking', 'FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes',
               'Depression', 'HeadInjury', 'Hypertension', 'MemoryComplaints', 'BehavioralProblems', 'Confusion', 'Disorientation',
               'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness', 'Diagnosis']


continuous_vars = [col for col in df.columns if col not in binary_vars]

plt.figure(figsize=(10, 20))  # Ajusta el tamaño según la cantidad de variables
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(len(continuous_vars), 1, i)
    sns.boxplot(x=df[var])
    plt.title(var)

plt.tight_layout()  # Ajusta los subplots para que no se sobrepongan
plt.show()
```

Se puede comprobar que las variables están todas bien distribuidas, ya que la media está más o menos centrada en todas y que no presentan outliers. Por lo que ya está toda la base de datos lista para el análisis.

#Análisis de correlación:

Para continuar, se verá qué variables están más correlacionadas para ver si hay alguna correlación inesperada e interesante para estudiar. Esto no es más que un análisis 'a ojo'.

```python
plt.figure(figsize=(12, 10))

# Mapa de calor con etiquetas rotadas
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 8})

plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x
plt.yticks(rotation=0)  # Mantener las etiquetas del eje y horizontales

plt.tight_layout()  # Ajusta el layout para que no se superpongan
plt.show()
```

Las 5 correlaciones más altas:

```python
import pandas as pd
import numpy as np

# Deshacemos la simetría obteniendo solo la mitad superior de la matriz, excluyendo la diagonal
corr_pairs = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Desempaquetamos en una tabla ordenada
corr_pairs_unstacked = corr_pairs.unstack().reset_index()
corr_pairs_unstacked.columns = ['Variable1', 'Variable2', 'Correlation']

# Filtramos los valores NaN (que son las combinaciones excluidas de la mitad inferior)
corr_pairs_filtered = corr_pairs_unstacked.dropna()

# Ordenamos por valor absoluto de la correlación
corr_pairs_filtered['AbsCorrelation'] = corr_pairs_filtered['Correlation'].abs()
sorted_corr_pairs = corr_pairs_filtered.sort_values(by='AbsCorrelation', ascending=False)

# Mostramos las correlaciones más altas (por ejemplo, las 5 primeras)
print(sorted_corr_pairs.head(5))
```

Las correlaciones más altas se encuentran entre 0.2 y 0.4, por lo que no se pueden considerar muy altas. Es verdad, que todas se relacionan con diagnosis, que es la variable objetivo, por lo que se pueden hacer unas primeras intuiciones.

A priori, se ve que la variable que más afecta (con una correlación negativa) es FunctionalAssessment. Esta medía la capacidad de hacer actividades cotidianas, cuanto más alta es hay más probabilidad de que presente Alzheimer. Ocurre lo mismo con la segunda más alta ADL, que tiene que ver con las actividades que se realizan como bañarse, vestirse... también afecta negativamente.
